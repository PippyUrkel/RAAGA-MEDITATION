```python
import torchaudio
from audiocraft.models import musicgen
from audiocraft.utils.notebook import display_audio
```

```python
model = musicgen.MusicGen.get_pretrained('facebook/musicgen-large', device='cuda')
model.set_generation_params(duration=30)
```

```python
import soundfile as sf

res = model.generate([
    'classical indian music', 
    'classic reggae track with an electronic guitar solo'],
    progress=True)

# Get the first generated sample (or loop through if needed)
audio = res[0].cpu().numpy()  # shape: (1, num_samples, channels)
audio = audio.squeeze()       # removes any 1-sized dimensions => (num_samples, channels)

sf.write('output_audio.wav', audio, 32000)
```
